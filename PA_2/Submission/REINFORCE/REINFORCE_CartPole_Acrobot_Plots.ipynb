```python
# Import the libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# CartPole-v1 plots

# Read reward mean data from csv file
file_name_mean_N = "REINFORCE_CartPole_v1_mean.csv"
file_name_std_N = "REINFORCE_CartPole_v1_std.csv"
file_name_mean_B = "REINFORCEB_CartPole_v1_mean.csv"
file_name_std_B = "REINFORCEB_CartPole_v1_std.csv"

# Read data from csv file
mean_N = np.loadtxt(file_name_mean_N, delimiter=',')
std_N = np.loadtxt(file_name_std_N, delimiter=',')
mean_B = np.loadtxt(file_name_mean_B, delimiter=',')
std_B = np.loadtxt(file_name_std_B, delimiter=',')

# Plot the reward mean
plt.figure()
plt.plot(mean_N, label='REINFORCE')
plt.fill_between(range(len(mean_N)), mean_N-std_N, mean_N+std_N, alpha=0.2)
plt.plot(mean_B, label='REINFORCE with baseline')
plt.fill_between(range(len(mean_B)), mean_B-std_B, mean_B+std_B, alpha=0.2)
plt.xlabel('Episode')
plt.ylabel('Mean of Average Reward')
plt.title('REINFORCE CartPole-v1')
plt.legend()
plt.grid()
plt.show()
```
